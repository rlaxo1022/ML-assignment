## compute distance
- feature : $n \times m$, center : $1 \times m$, distance : $n \times 1$
- $n$ : number of data, $m$ : number of features

```python
def compute_distance(feature, center):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    
    
    # m = 2
    # m = x에 대한 centroid, y에 대한 centroid
    # print(feature.shape[0]-1)
    # center = np.zeros((1, feature.shape[1]))
    # distance = []
    # # # for i in range (0, feature.shape[0]-1):
    # distance = np.vstack(np.sqrt(sum(np.power(feature - center, 2))))
    
    # center=center[:,np.newaxis,:]
    # distance=np.sqrt(np.sum(np.square(center-feature),axis=2))

    distance = np.expand_dims(feature, axis=1) - center
    distance = np.square(distance)
    distance = np.sqrt(np.sum(distance, axis=2))
    

           

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return distance
```
